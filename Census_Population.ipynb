```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#read csv file
census = "PEP_2016_PEPAGESEX_with_ann_columnedit.csv"
```

```python

# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
popu_df = pd.read_csv(census, encoding = "ISO-8859-1")
popu_df= popu_df.drop_duplicates()

popu_df
```

```python
popu_df.columns.values.tolist()
```

```python
# chnage the column title to a meaningful title

popu_df.rename(columns={'Geography': 'States'}, inplace=True)

popu_df
```

```python

# st = popu_df["States"]
x = popu_df["Id"]
y = popu_df["Population Estimate (as of July 1) - 2010 - Both Sexes; Total"]
z= popu_df["Population Estimate (as of July 1) - 2011 - Both Sexes; Total"]
k = popu_df["Population Estimate (as of July 1) - 2012 - Both Sexes; Total"]
```

```python
ax = plt.subplot(111)
w = 0.3

ax.bar(x, y,width=w,color='b',align='center')
ax.bar(x, z,width=w,color='g',align='center')
ax.bar(x, k,width=w,color='r',align='center')

ax.autoscale(tight=True)

plt.show()
```
